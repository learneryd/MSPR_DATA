# Analyse du SMIC Brut Horaire

Ce notebook traite le fichier `valeurs_annuelles.csv`. Nous allons supprimer les lignes inutiles, renommer les colonnes, et créer un graphique du SMIC brut horaire par an.


```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
 
import pandas as pd

# Lire le fichier CSV en spécifiant le séparateur comme point-virgule (ou autre si nécessaire)
file_path = 'valeurs_annuelles.csv'
data = pd.read_csv(file_path, sep=';', skiprows=3, header=None)

# Afficher les premières lignes du DataFrame pour vérifier les données initiales
print("Données initiales :")
print(data.head())

# Afficher les noms des colonnes pour vérifier leur structure
print("\nColonnes initiales :")
print(data.columns)
```

```python

# Renommer les colonnes en fonction de votre besoin
data.columns = ['année', 'SMIC brut horaire', 'Status']

# Afficher les noms des colonnes pour vérifier leur structure
print("\nColonnes après renommage :")
print(data.columns)
```

```python
# Vérifier que les colonnes nécessaires existent
if 'année' in data.columns and 'SMIC brut horaire' in data.columns:
    # Créer un graphique
    plt.figure(figsize=(12, 6))
    plt.plot(data['année'], data['SMIC brut horaire'], marker='o', linestyle='-', color='b')
    plt.title('SMIC Brut Horaire par An')
    plt.xlabel('Année')
    plt.ylabel('SMIC Brut Horaire')
    plt.grid(True)
    plt.xticks(rotation=45)  # Rotation des étiquettes de l\'axe des x pour une meilleure lisibilité
    plt.tight_layout()  # Ajuster la disposition pour s\'adapter aux étiquettes

    plt.gca().invert_xaxis()
    # Afficher le graphique
    plt.show()
else:
    print("Les colonnes nécessaires 'année' et 'smic brut horaire' ne sont pas présentes dans les données.")
```
