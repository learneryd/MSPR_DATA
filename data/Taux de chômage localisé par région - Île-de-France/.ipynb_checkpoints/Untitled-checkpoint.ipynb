# Taux de chomage en Ile De France   2012 - 2024

```python
# Importer les bibliothèques nécessaires
import pandas as pd
import matplotlib.pyplot as plt
```

```python

# Lire le fichier CSV en spécifiant le séparateur comme point-virgule et garder les données à partir de la ligne 8
file_path = 'valeurs_trimestrielles.csv'
data = pd.read_csv(file_path, sep=';', skiprows=7, usecols=[0, 1], header=None)

# Renommer les colonnes
data.columns = ['anne_trimestre', 'taux_chomage']
```

```python

# Afficher les premières lignes du DataFrame pour vérifier les données initiales
print("Données initiales :")
print(data.head())
```

```python

# Convertir les colonnes en numériques et nettoyer les données
# Convertir 'anne_trimestre' en numérique
data['anne_trimestre'] = pd.to_numeric(data['anne_trimestre'].str.extract(r'(\d{4})')[0], errors='coerce')

# Assurer que 'taux_chomage' est bien en format string, puis enlever les guillemets et convertir en numérique
data['taux_chomage'] = data['taux_chomage'].astype(str).str.replace('"', '', regex=False)
data['taux_chomage'] = pd.to_numeric(data['taux_chomage'], errors='coerce')

# Vérifier que les colonnes nécessaires existent
if 'anne_trimestre' in data.columns and 'taux_chomage' in data.columns:
    # Créer un graphique
    plt.figure(figsize=(12, 6))
    plt.plot(data['anne_trimestre'], data['taux_chomage'], marker='o', linestyle='-', color='b')
    plt.title('Taux de Chômage par Année')
    plt.xlabel('Année')
    plt.ylabel('Taux de Chômage')
    plt.grid(True)
    plt.xticks(rotation=45)  # Rotation des étiquettes de l'axe des x pour une meilleure lisibilité
    plt.tight_layout()  # Ajuster la disposition pour s'adapter aux étiquettes
    
    # Inverser l'ordre des années sur l'axe des x
    
    
    # Afficher le graphique
    plt.show()
else:
    print("Les colonnes nécessaires 'anne_trimestre' et 'taux_chomage' ne sont pas présentes dans les données.")
```
